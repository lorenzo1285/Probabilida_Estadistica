```python
import statsmodels as sm
import scipy.stats
```

**Establecer Hipotesis Nula y Alternativa**

S_1 = 2.79 (AMX)

S_2 = 4.11 (GEO)

H_0 : S_1 / S_2 = 1

H_1 : S_2 / S_1 > 1

```python
S_1 = 2.79 # Varianza (AMX)
S_2 = 4.11 # Varianza (GEO)
n_1 = 42 - 1 # Grados de libertad
n_2 = 42 - 1 # Grados de libertad
a = .01 # Nivel de Significancia
q = 1 - a  
```

```python
F =  S_2/S_1 #Estadistico de prueba
F # F calculado 
```

```python
F_c = scipy.stats.f.ppf(q , dfn= n_1, dfd= n_2) 
F_c # F critica
```

```python
if F > F_c:
  print('Se rechaza la hipotesis Nula')
else:
  print('Se acepta la hipotesis Nula')
```
