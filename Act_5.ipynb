2.	Una empresa empaca arroz en dos presentaciones, una bolsa que contiene más o menos 1 kilo de arroz y la otra presentación es en bultos de arroz de 20 kilos. Una muestra aleatoria de 35 bolsas produjo mediciones de peso con un promedio de 1.01 kilos y una desviación estándar de .18 kilos.

•	Elabora un intervalo de confianza de 99% para el peso promedio de las bolsas que empaca esta empresa

•	Responde: ¿Qué significa la frase “confianza del 99%”?

•	El departamento de control de calidad de esta empresa piensa que el peso de las bolsas debe ser en promedio de 1 kilo. Responde: ¿Debe preocupar al departamento de control de calidad el intervalo de confianza? Justifica tu respuesta. 

```python
n = 35
std = .18
media = 1.01
z = stats.norm.ppf(q = 0.995)
```

```python
z
```

```python
margen_error = z * (std/math.sqrt(n))
IC = [media - margen_error,
      media + margen_error]
print('Intervalo de Confianza del 99% de la Media')
IC 
```

3.	Para el caso de los bultos de arroz de 20 kilos, se tomó una muestra aleatoria de 10 bultos, fueron pesados y se obtuvieron las siguientes mediciones 18, 19, 21, 20, 20, 19.5, 20.5, 21.5, 18, 20.

•	Elabora un intervalo de confianza del 95% para el peso promedio de los bultos de arroz que empaca esta empresa.

•	Responde: ¿Qué significa la frase “confianza del 95%”?

•	El departamento de control de calidad de esta empresa piensa que el peso de los bultos de arroz debe ser en promedio de 20 kilos. Responde: ¿Debe preocupar al departamento de control de calidad el intervalo de confianza obtenido? Justifique su respuesta.


```python
n_1 = 10
muestra = pd.DataFrame({'Bultos':[18,19,21,20,20,19.5,20.5,21.5,18,20]})
media_muestral = muestra.Bultos.mean()
t = stats.t.ppf(q = 0.975, df=9)
std_muestral = muestra.Bultos.std(ddof=1)
```

```python
sigma = std_muestral/math.sqrt(n_1)
margen_error_1 = t*sigma
```

```python
IC_1 = (media_muestral - margen_error_1,
        media_muestral + margen_error_1)
print('Intervalo de Confianza del 95% de la media muestral')
IC_1
```

4.	José Francisco un escritor reconocido acaba de terminar una primera versión de su libro de 700 páginas. José capturo su libro en un procesador de textos y está interesado en conocer el número promedio de errores de captura por página, pero no quiere leer todo el documento. José tomo una muestra de 40 páginas seleccionadas de manera aleatoria y encontró que el promedio de errores de captura por página fue 4.3 y la desviación estándar de la muestra fue 1.2 errores por página.

•	Calcula el error estándar estimado de la media

•	Calcula un intervalo de confianza del 90% para el número promedio verdadero de errores por página en su libro


```python
n_2 = 40
media_muestral_2 = 4.3
std_muestral_2 = 1.2
sigma_2 = std_muestral_2/math.sqrt(n_2)
```

```python
IC_2 = stats.t.interval(alpha= 0.95,
                 df= 39,
                 loc= media_muestral_2,
                 scale= sigma_2)
```

```python
print('Error estandar =', sigma_2)
```

```python
print('Intervalo de Confinza del 90% de la Media Muestral')
IC_2
```

5.	Una operación de ensamblaje en una planta de automóviles requiere aproximadamente un mes de capacitación con el fin de que un nuevo empleado alcance la eficiencia máxima en el ensamblaje de un automóvil. Se sugirió un nuevo método de capacitación y se llevó a cabo una prueba para comparar el nuevo método con el procedimiento estándar. Se capacitó a dos grupos de nueve empleados durante tres semanas, uno con el nuevo método y otro siguiendo el procedimiento de capacitación estándar.  Se registró el tiempo requerido para que cada empleado ensamblara el automóvil al final de las tres semanas, los resultados fueron los siguientes:

Procedimiento estándar			Nuevo procedimiento

X1 = 35.22          X2 = 31.56

S1 = 4.9441					S2 = 4.4752



```python
n_3 = 9
X1 = 35.22
X2 = 31.56
S1 = 4.9441
S2 = 4.4752 
z_3 = stats.t.ppf(q = 0.975, df= 8)
```

```python
z_3
```

```python
sig_S1 = S1/math.sqrt(n_3)
sig_S2 = S2/math.sqrt(n_3)
```

```python
sig_S1, sig_S2
```

```python
margen_error_3 = z_3 * (S1/math.sqrt(n_3))
metodo_1 = (X1 - margen_error_3,
            X1 + margen_error_3)
print('Intervalo de Confianza del 95% de la Media del Procedimiento Estandar')
metodo_1 
```

```python
margen_error_4 = z_3 * (S2/math.sqrt(n_3))
metodo_2 = (X2 - margen_error_4,
            X2 + margen_error_4)
print('Intervalo de Confianza del 95% de la Media del Procedimiento Nuevo')
metodo_2 
```

```python
IC_3 = stats.t.interval(alpha= 0.975,
                 df= 8,
                 loc= X1,
                 scale= sig_S1)
print('Intervalo de Confianza del 95% de la Media del Procedimiento Estandar')
IC_3
```

```python
IC_4 = stats.t.interval(alpha= 0.975,
                 df= 8,
                 loc= X2,
                 scale= sig_S2)
print('Intervalo de Confianza del 95% de la Media del Procedimiento Nuevo')
IC_4
```

```python
metodos = pd.DataFrame({'Metodo_Estandar': IC_3, 'Metodo_Nuevo':IC_4})
metodos
```

```python
diferencia = metodos.Metodo_Estandar - metodos.Metodo_Nuevo
print('Diferencia en Segundos')
diferencia
```

```python
print('Diferencia Porcentual')
diferencia / metodos.Metodo_Estandar
```
